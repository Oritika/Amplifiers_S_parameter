```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import scipy
import os
```

```python
files = ['/xmmscratch/singh/Excercise/Chan_1.csv', '/xmmscratch/singh/Excercise/Chan_2.csv',
         '/xmmscratch/singh/Excercise/Chan_3.csv', '/xmmscratch/singh/Excercise/Chan_4.csv', 
         '/xmmscratch/singh/Excercise/Chan_5.csv', '/xmmscratch/singh/Excercise/Chan_11.csv', 
         '/xmmscratch/singh/Excercise/Chan_12.csv', '/xmmscratch/singh/Excercise/Chan_13.csv', 
         '/xmmscratch/singh/Excercise/Chan_14.csv', '/xmmscratch/singh/Excercise/Chan_15.csv']
```

```python
all_files = os.listdir('Excercise')
```

```python
for one_file in all_files:
    data = np.loadtxt(os.path.join('Excercise', one_file),skiprows= 7, delimiter=',')
    fig,ax1 = plt.subplots()
    ax2 = ax1.twinx()
    plt.grid(True)
    plot_1 = ax1.plot(data[:,0]/1e6,data[:,5])
    plot_2 = ax2.plot(data[:,0]/1e6,data[:,1])
    ax1.set_ylabel('S21(MAG) (dB)')
    ax2.set_ylabel('S11(MAG) (dB)')
    plt.xlabel('Frequency (Hz)')
    #plt.ylabel('Magnitude (dB)')
    
```

```python
plt.figure(figsize=(12, 8))
for one_file in all_files:
    data = np.loadtxt(os.path.join('Excercise', one_file),skiprows= 7, delimiter=',')
    plt.subplot(2, 2, 1)
    plt.plot(data[:,0]/1e6,data[:,1])
    plt.title('S11')
    plt.legend(all_files)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude (dB)')

    plt.subplot(2, 2, 2)
    plt.plot(data[:,0]/1e6,data[:,3])
    plt.title('S12')
    plt.legend(all_files)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude (dB)')

    plt.subplot(2, 2, 3)
    plt.plot(data[:,0]/1e6,data[:,5])
    plt.title('S21')
    plt.legend(all_files)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude (dB)')

    plt.subplot(2, 2, 4)
    plt.plot(data[:,0]/1e6,data[:,7])
    plt.title('S22')
    plt.legend(all_files)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude (dB)')

plt.savefig('/xmmscratch/singh/Amp o_p images/S_paras.png')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for one_file in all_files:
    data = np.loadtxt(os.path.join('Excercise', one_file),skiprows= 7, delimiter=',')
    
    plt.subplot(1, 3, 1)
    plt.plot(np.corrcoef(data[:,1],data[:,3]))
    plt.legend(all_files)
    plt.title('correlation between S11 and S12')
    
    plt.subplot(1, 3, 2)
    plt.plot(np.corrcoef(data[:,1],data[:,5]))
    plt.legend(all_files)
    plt.title('correlation between S11 and S21')

    plt.subplot(1, 3, 3)
    plt.plot(np.corrcoef(data[:,1],data[:,7]))
    plt.legend(all_files)
    plt.title('correlation between S11 and S22')
plt.savefig('/xmmscratch/singh/Amp o_p images/corr_1.png')    
plt.tight_layout()
plt.show()
```
